Apply method

```python
# In this method we can create a function and then apply it to execute operation
```

```python
def marks(x):
    return x/2      # x//2 for integer value
```

```python
df1['Half_marks']=df1['Marks'].apply(marks)
```

```python
df1
```

```python
# similar result using lambda function
df1['Half_marks']=df1['Marks'].apply(lambda x:x//2)
```

```python
df1
```

```python
# length of strings of column
df1['Name'].apply(len)
```
